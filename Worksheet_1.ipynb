```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
# Compute the inverse of A
A_inv = np.linalg.inv(A)

# Compute A * A^-1
result = np.dot(A, A_inv)

# Verify if the result is an identity matrix
identity_matrix = np.eye(2)  # Identity matrix of size 2x2
print("1. A * A^-1:")
print(result)
print("Is A * A^-1 equal to the identity matrix?", np.allclose(result, identity_matrix))

# Compute AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\n2. AB and BA:")
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Is AB equal to BA?", np.allclose(AB, BA))  # This should return False

# Compute (AB)^T and B^T * A^T
AB_transpose = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))

print("\n3. (AB)^T and B^T * A^T:")
print("(AB)^T:")
print(AB_transpose)
print("B^T * A^T:")
print(BT_AT)
print("Is (AB)^T equal to B^T * A^T?", np.allclose(AB_transpose, BT_AT))

# Define matrix A and vector B
coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3,  1, -1]])
constants = np.array([-1, -3, 9])

# Solve for X using np.linalg.inv
A_inv = np.linalg.inv(coeff_matrix)
solution = np.dot(A_inv, constants)

print("\n4. Solution to the system of equations using inverse method:")
print("Solution (x, y, z):", solution)


# Solve for X directly using np.linalg.solve
solution_direct = np.linalg.solve(coeff_matrix, constants)

print("\n5. Solution to the system of equations using np.linalg.solve:")
print("Solution (x, y, z):", solution_direct)
```
