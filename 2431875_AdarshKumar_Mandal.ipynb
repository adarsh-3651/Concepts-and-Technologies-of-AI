####  Task- 3- Outlier Detection:

```python
#  1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
#  2. Define outliers using the 1.5 × IQR rule.
#  3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
#  in a different color.

import pandas as pd
import matplotlib.pyplot as plt

# Filter dataset for South Asian countries
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Define function to calculate outliers using 1.5 × IQR rule
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Identify outliers for Score and GDP per Capita
score_outliers = detect_outliers(south_asia_df, 'score')
gdp_outliers = detect_outliers(south_asia_df, 'Log GDP per capita')

# Combine all outliers
outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

# Scatter Plot: GDP per Capita vs. Score with outliers highlighted
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'],
            color='blue', label='Normal Countries', alpha=0.7, edgecolor='black')
plt.scatter(outliers['Log GDP per capita'], outliers['score'],
            color='red', label='Outliers', alpha=1, edgecolor='black')
plt.title('Outlier Detection: GDP per Capita vs. Score (South Asia)', fontsize=16)
plt.xlabel('GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Display Outliers
print("Outliers in South Asia based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])

# Interpretation:
# 1. Afghanistan is likely identified as a low outlier due to its low Score and GDP per Capita.
# 2. High outliers might include countries with exceptional economic performance relative to their regional peers.
# 3. Outliers skew the average Score and GDP, making South Asia appear less cohesive.


#  4. Discuss the characteristics of these outliers and their potential impact on regional averages.

# Characteristics:
#   Outliers like Afghanistan (low Score and GDP) represent countries with extreme economic and social instability.
#   High-performing countries, if any, might exhibit unusually high GDP per Capita or Score compared to the regional average.

# Impact on Regional Averages:
#   Low outliers significantly reduce the average Score and GDP per Capita for South Asia.
#   Outliers can distort statistical measures like the mean, but median-based metrics are less affected.

```

####  Task- 4- Exploring Trends Across Metrics:

```python
#  1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
#  {pearson correlation} with the Score for South Asian countries.
#  2. Create scatter plots with trendlines for these metrics against the Score.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/Concept of Technologies and AI/WHR-2024-5CS037.csv')

# Define South Asian countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter dataset for South Asian countries
south_asia_df = df[df['Country name'].isin(south_asian_countries)]

# Calculate Pearson Correlation
correlation_freedom = south_asia_df[['score', 'Freedom to make life choices']].corr(method='pearson').iloc[0, 1]
correlation_generosity = south_asia_df[['score', 'Generosity']].corr(method='pearson').iloc[0, 1]

print(f"Correlation of Score with Freedom to Make Life Choices: {correlation_freedom}")
print(f"Correlation of Score with Generosity: {correlation_generosity}")

# Scatter Plot with Trendline: Freedom to Make Life Choices vs. Score
plt.figure(figsize=(10, 6))
sns.regplot(data=south_asia_df, x='Freedom to make life choices', y='score',
            line_kws={"color": "red"}, scatter_kws={"alpha": 0.7})
plt.title('Score vs. Freedom to Make Life Choices (South Asia)', fontsize=16)
plt.xlabel('Freedom to Make Life Choices', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Scatter Plot with Trendline: Generosity vs. Score
plt.figure(figsize=(10, 6))
sns.regplot(data=south_asia_df, x='Generosity', y='score',
            line_kws={"color": "orange"}, scatter_kws={"alpha": 0.7})
plt.title('Score vs. Generosity (South Asia)', fontsize=16)
plt.xlabel('Generosity', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Interpretation:
# Compare the Pearson correlation values to identify the strongest and weakest relationships.
# Stronger relationships will show a clear positive or negative trend in the scatter plot with the trendline.
# Weak relationships will show a scattered distribution with a near-flat trendline.


#  3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
#  South Asian countries.

# Strongest Relationship: Freedom to Make Life Choices shows a strong positive correlation with Score, indicating that greater personal freedom significantly boosts happiness in South Asia.
# Weakest Relationship: Generosity has a weak positive correlation, suggesting it has little direct impact on happiness scores.
# Trendline: The scatter plot for Freedom shows a clear upward trend, while Generosity appears scattered with no strong trend.
# Conclusion: Autonomy matters more for happiness in South Asia than generosity.
```

####  Task- 5- Gap Analysis:

```python
# 1. Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score
#  for each South Asian country.
#  2. Rank the South Asian countries by this gap in both ascending and descending order.
#  3. Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.
#  4. Analyze the reasons behind these gaps and their implications for South Asian countries.
# Analysis
# Top Positive Gaps
#   Countries: Those with high GDP per Capita but relatively low Happiness Scores.
#   Reason:
#     Economic disparity where high GDP benefits only a small segment of the population.
#     Weakness in other happiness factors such as social support, freedom, or health.
# Top Negative Gaps
#   Countries: Those with low GDP per Capita but higher Happiness Scores.
#   Reason:
#     Socio-cultural factors like strong community support or high levels of generosity.
#     Greater happiness despite economic limitations, indicating resilience or effective non-monetary well-being.



# Calculate GDP-Score Gap
south_asia_df['GDP-Score Gap'] = south_asia_df['Log GDP per capita'] - south_asia_df['score']

# Rank countries by GDP-Score Gap
ascending_rank = south_asia_df.sort_values(by='GDP-Score Gap', ascending=True)
descending_rank = south_asia_df.sort_values(by='GDP-Score Gap', ascending=False)

# Display top 3 positive and negative gaps
top_positive_gap = descending_rank.head(3)
top_negative_gap = ascending_rank.head(3)

print("Top 3 Countries with Largest Positive GDP-Score Gap:")
print(top_positive_gap[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])

print("\nTop 3 Countries with Largest Negative GDP-Score Gap:")
print(top_negative_gap[['Country name', 'Log GDP per capita', 'score', 'GDP-Score Gap']])

# Bar chart: Positive and Negative Gaps
plt.figure(figsize=(12, 6))

# Positive Gaps
plt.subplot(1, 2, 1)
plt.bar(top_positive_gap['Country name'], top_positive_gap['GDP-Score Gap'], color='green', alpha=0.8)
plt.title('Top 3 Positive GDP-Score Gaps (South Asia)', fontsize=14)
plt.xlabel('Country name', fontsize=12)
plt.ylabel('GDP-Score Gap', fontsize=12)

# Negative Gaps
plt.subplot(1, 2, 2)
plt.bar(top_negative_gap['Country name'], top_negative_gap['GDP-Score Gap'], color='red', alpha=0.8)
plt.title('Top 3 Negative GDP-Score Gaps (South Asia)', fontsize=14)
plt.xlabel('Country', fontsize=12)

plt.tight_layout()
plt.show()
```
